```python
# ROEとROICの値を取得
# FinancialsはROEを使用
conn = sqlite3.connect(DATA_DIR / "MSCI_KOKUSAI" / "MSCI_KOKUSAI_Factset_Financials.db")

df = pd.read_sql(
    "SELECT * FROM all_data WHERE variable IN ('FF_ROIC', 'FF_ROE')", con=conn
).assign(date=lambda row: pd.to_datetime(row["date"]))
df = pd.pivot(
    df, index=["date", "P_SYMBOL"], columns="variable", values="value"
).reset_index()

# df_return = pd.read_sql("SELECT * FROM Return", con=conn)
# display(df_return)

# 銘柄のweight, market value, gics sectorを取得
conn = sqlite3.connect(
    DATA_DIR / "MSCI_KOKUSAI" / "MSCI_KOKUSAI_BPM_and_Factset_Code.db"
)

df_weight_and_sector = pd.read_sql(sql="SELECT * FROM all_data", con=conn)[
    ["date", "P_SYMBOL", "GICS Sector", "Mkt Value", "Weight (%)"]
].assign(date=lambda row: pd.to_datetime(row["date"]))

# merge
df_merged = (
    pd.merge(df, df_weight_and_sector, on=["date", "P_SYMBOL"], how="left")
    # FinancialsはROEを使用
    .assign(
        ROIC=lambda dfx: np.where(
            dfx["GICS Sector"] == "Financials", dfx["FF_ROE"], dfx["FF_ROIC"]
        )
    )
    .dropna(subset=["Weight (%)"], ignore_index=True)
    # FF_ROE, FF_ROICのカラムを削除
    .drop(columns=["FF_ROE", "FF_ROIC"])
)
display(df_merged)
```
